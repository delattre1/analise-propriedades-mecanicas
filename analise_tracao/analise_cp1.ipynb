```python
import pandas as pd
import matplotlib.pyplot as plt
import math
from math import pi
import numpy as np
```

```python
def calcula_area_secao_transv(diametro):
    raio = diametro/2
    return (raio**2)*math.pi

def calcula_tensao(forca, sec_transv):
    return forca/sec_transv
```

```python
df_corpo_prova_1 = pd.read_excel("5A.xlsx", engine="openpyxl")
df_corpo_prova_2 = pd.read_excel("5B.xlsx", engine="openpyxl")
df_corpo_prova_3 = pd.read_excel("5C.xlsx", engine="openpyxl")
```

```python
df_corpo_prova_1[0:5]
```

```python
#grafico tensao X deformacao - até 1%
diametro_cp1 = df_corpo_prova_1['Diâmetro (mm)'][0]
diametro_cp2 = df_corpo_prova_2['Diâmetro (mm)'][0]
diametro_cp3 = df_corpo_prova_3['Diâmetro (mm)'][0]

tam_final_barra_cp1 = df_corpo_prova_1['Lf (mm)'][0]
tam_final_barra_cp2 = df_corpo_prova_2['Lf (mm)'][0]
tam_final_barra_cp3 = df_corpo_prova_3['Lf (mm)'][0]
l0 = 50 #tamanho inciial da barra

sec_transv1 = calcula_area_secao_transv(diametro_cp1)
tensao1 = calcula_tensao(df_corpo_prova_1["Force (N)"], sec_transv1)
df_corpo_prova_1["Tensão [MPa]"] = tensao1

sec_transv2 = calcula_area_secao_transv(diametro_cp2)
tensao2 = calcula_tensao(df_corpo_prova_2["Force (N)"], sec_transv2)
df_corpo_prova_2["Tensão [MPa]"] = tensao2

sec_transv3 = calcula_area_secao_transv(diametro_cp3)
tensao3 = calcula_tensao(df_corpo_prova_3["Force (N)"], sec_transv3)
df_corpo_prova_3["Tensão [MPa]"] = tensao3

```
